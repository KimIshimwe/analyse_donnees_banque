```python
# créer des colonnes d'âge (<18, entre 18 et 25, entre 25 et 50 et plus de 50)
# Il faut d'abord créér une nouvelle colonne composée de valeurs vides
import numpy as np
df["tranche_age"] = np.nan
df.columns
df["tranche_age"] # il n'y a que des valeurs vides

```

```python
# Approche ( attribution des valeurs de la colonnes
df["tranche_age"][df["age"] <18] = "< 18"
df["tranche_age"][  (df["age"] >= 18) & (df["age"] <= 25) ] = " 18-25 "
df["tranche_age"][(df["age"] >= 25) & (df["age"] <= 50)] = "25-50"
df["tranche_age"][df["age"] > 50] = ">50"

```

```python
df.describe(include = "all")
df["tranche_age"].value_counts().plot(kind = "bar")
```

```python
# Quel est le solde moyen par tranche d'âge ?
df.groupby("tranche_age")["balance"].mean()
```

```python
df_plus_de_25 = df[df["age"] > 25] # création de la base des ind de plus de 25
df_plus_de_25.groupby("education")["age"].mean() # avoir la moyenne d'age des individus de plus de 25
```
